```python
%pylab inline
from scipy.fftpack import fft, fftfreq
#Carga los datos en un array
data= loadtxt('sparse_sample_monthrg.dat')
cor_data=[]
for row in data:
    if (row[3]!=-99.0):
        row[0]+=(row[1]/12)-(1/12)
        cor_data.append(row)
cor_data=array(cor_data)
plot(cor_data[:,0],cor_data[:,3])
```

```python
u=5
x_in=cor_data[0:u,0]
y_in=cor_data[0:u,3]
x_cua=linspace(cor_data[0,0],cor_data[len(cor_data[0:u,0])-1,0],u)
y_cua=0
for i in range(len(x_in)):
    b0=1
    for k in range (len(x_in)):
        if(i!=k):
            b1=((x_cua-x_in[k])/(x_in[i]-x_in[k]))
            b0=b0*b1
    y_cua = y_cua+b0*y_in[i]
    #plot(x_cua,b0*y_in[i])
plot(x_cua, y_cua)
scatter(x_in, y_in, alpha=1)  
#plot (x_in,y_in)
```
